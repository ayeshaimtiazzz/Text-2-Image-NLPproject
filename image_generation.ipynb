```python
from diffusers import DiffusionPipeline
import torch
import matplotlib.pyplot as plt

# Load the full SDXL 1.0 model
pipe = DiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    torch_dtype=torch.float16,
    variant="fp16"
)
pipe = pipe.to("cuda")
```

```python

from textblob import TextBlob

user_prompt = input("Enter a text prompt for image generation: ")
corrected_prompt = str(TextBlob(user_prompt).correct())

print(f"Corrected prompt: {corrected_prompt}")

# Now use corrected_prompt instead of user_prompt in pipe(...)
with torch.no_grad():
    image = pipe(corrected_prompt, num_inference_steps=25, guidance_scale=7.5).images[0]
plt.figure(figsize=(8, 8))
plt.imshow(image)
plt.axis("off")
plt.show()
```
